```python
from transformers import AutoModel, BertForSequenceClassification
import weightwatcher as ww

import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

import warnings
import logging
warnings.filterwarnings("ignore")
logging.getLogger("transformers").setLevel(logging.ERROR)

import matplotlib.pyplot as plt
```

```python
model = AutoModel.from_pretrained('bert-base-uncased')
```

```python
watcher = ww.WeightWatcher(model=model)
results = watcher.analyze()
```

```python
results
```

```python
results[['longname','alpha']]
```

```python
results['longname'].unique()[-10:]#checking last 10 layer names
```

```python
pattern_attention = r'encoder\.layer\.\d+\.attention\.output\.dense'
pattern_output = r'encoder\.layer\.\d+\.output\.dense'
```

```python
df_output_dense = results[results['longname'].str.contains(pattern_output, regex=True)]
df_output_dense['id'] = df_output_dense['longname'].str.extract(r'encoder\.layer\.(\d+)')
```

```python
df_output_dense
```

```python
import matplotlib.pyplot as plt

df = results
# Get alpha values and layer ids
alphas = df['alpha'].values
ids = df['layer_id'].values 

# Find the maximum alpha and its corresponding layer id
max_alpha_idx = alphas.argmax()
max_alpha = alphas[max_alpha_idx]
max_alpha_id = ids[max_alpha_idx]

# Plot alpha values
plt.figure(figsize=(10, 5))
plt.plot(ids, alphas, marker='o')
plt.xlabel('Layer ID')
plt.ylabel('Alpha')
plt.title('all layers')

# Annotate the peak value on the plot
for idx, alpha in zip(ids, alphas):
    plt.annotate(f'{alpha:.2f}', 
                 (idx, alpha), 
                 textcoords="offset points", 
                 xytext=(-10,10), 
                 ha='center', 
                 fontsize=5)
```

# Combined graphs

```python
def annotate_axes(ax, text, fontsize=18):
    ax.text(0.5, 0.5, text, transform=ax.transAxes,
            ha="center", va="center", fontsize=fontsize, color="darkgrey")
fig, axd = plt.subplot_mosaic([['upper', 'upper'],
                               ['lower left', 'lower right']],
                              figsize=(10, 7), layout="constrained")
for k in axd:
    if k == 'upper':        
        df = results
        # Get alpha values and layer ids
        alphas = df['alpha'].values
        ids = df['layer_id'].values 

        axd[k].plot(ids, alphas, marker='o')
        axd[k].set_xlabel('Layer ID')
        axd[k].set_ylabel('Alpha')
        axd[k].set_title('all layers')

        # Annotate the peak value on the plot
        for idx, alpha in zip(ids, alphas):
            axd[k].annotate(f'{alpha:.2f}', 
                         (idx, alpha), 
                         textcoords="offset points", 
                         xytext=(-10,10), 
                         ha='center', 
                         fontsize=7)
#         axd[k].legend()
    if k=='lower left':
        df = df_output_dense
        # Get alpha values and layer ids
        alphas = df['alpha'].values
        ids = df['id'].values 

        axd[k].plot(ids, alphas, marker='o')
        axd[k].set_xlabel('Encoder layer ID')
        axd[k].set_ylabel('Alpha')
        axd[k].set_title('encoder.layer.x.output.dense')

        # Annotate the peak value on the plot
        for idx, alpha in zip(ids, alphas):
            axd[k].annotate(f'{alpha:.2f}', 
                         (idx, alpha), 
                         textcoords="offset points", 
                         xytext=(-10,10), 
                         ha='center', 
                         fontsize=7)
            
    if k=='lower right':   
        df = df_output_attention_dense
        # Get alpha values and layer ids
        alphas = df['alpha'].values
        ids = df['id'].values 

        axd[k].plot(ids, alphas, marker='o')
        axd[k].set_xlabel('Encoder layer ID')
        axd[k].set_ylabel('Alpha')
        axd[k].set_title('encoder.layer.x.attention.output.dense')

        # Annotate the peak value on the plot
        for idx, alpha in zip(ids, alphas):
            axd[k].annotate(f'{alpha:.2f}', 
                         (idx, alpha), 
                         textcoords="offset points", 
                         xytext=(-10,10), 
                         ha='center', 
                         fontsize=7)
        
        
        
fig.suptitle('Bert Alphas across layers ')
plt.show()
```
